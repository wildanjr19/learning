```python
import pandas as pd
```

```python
# load data train
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
# load data test
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
# melihat tipe data
df_train.info()
```

```python
# melihat analisis deskriptif data train
df_train.describe(include='all')
```

```python
# memeriksa missing value
missing_value = df_train.isnull().sum()
missing_value[missing_value > 0]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# memeriksa outlier pada data

for feature in df_train.columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df_train[feature])
    plt.title('Boxplot Data Train')
    plt.show()
```

Pada kasus ini kita akan menangani outlier dengan menghapusnya. Kita menggunakan metode IQR untuk mengidentifikasi outlier kemudian menghapusnya

```python
# identifikasi outlier
Q1 = df_train.quantile(0.25)
Q3 = df_train.quantile(0.75)
IQR = Q3 - Q1
```

```python
# filter dataframe agar hanya menyimpan baris yang tidak mengandung outler / menghapus baris-baris yang mengandung outlier
condition = ~((df_train < (Q1 - 1.5*IQR)) | (df_train > (Q3 + 1.5*IQR))).any(axis=1)

df = df_train.loc[condition, df_train.columns]
```

```python
for feature in df.columns:
    plt.figure(figsize=(8,7))
    sns.boxplot(x=df[feature])
    plt.title('Boxplot data tanpa outlier')
    plt.show()
```

```python
# melakukan standarisasi pada data
from sklearn.preprocessing import StandardScaler
```

```python
# memastikan kembali hanya tipe data numerik yang akan di Standarisasi
numeric_features = df.select_dtypes(include=['number']).columns
numeric_features
```

```python
# proses standarisasi
scaler = StandardScaler()

df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
# mengecek duplikasi data
duplicates = df.duplicated()

print('Baris Duplikat')
print(df[duplicates])
```

Visualisasi Data

```python
# menghitung jumlah variabel
num_vars = df.shape[1]
```

```python
# menenetukan jumlah baris dan kolom untuk grid subplot
n_cols = 4 # jumlah kolom yang diinginkan untuk subplot
n_rows = -(-num_vars // n_cols) # jumlah baris yang diinginkan
```

```python
# membuat subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))

# flatten axes array untuk memudahkan iterasi jila diperlukan
axes = axes.flatten()


# plot setiap variabel
for i, column in enumerate(df.columns):
    df[column].hist(ax=axes[i], bins=20, edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequecy')

# menghapus subplot yang tidak terpakai (jika ada)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# menyesuaikan layout agar lebih rapi
plt.tight_layout()
plt.show()
```

```python
# melihat korelasi tiap variabel (feature) terhadap target
target_corr = df.corr()['FloodProbability']
```

```python
# mengurutkan nilai korelasi
target_corr_sorted = target_corr.abs().sort_values(ascending=False)
```

```python
# visualisasi korelasi
plt.figure(figsize=(10, 6))
target_corr_sorted.plot(kind='bar')
plt.title('Korelasi dengan FloodProbability')
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficients')
plt.show()
```

Splitting Data

```python
# deklarasi variabel X dan Y
X = df.drop(columns=['FloodProbability'])
y= df['FloodProbability']
```

```python
# split data menjadi train dan test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# melihat hasil split data
print('Jumlah Data', len(X))
print('Jumlah Data Latih', len(x_train))
print('Jumlah Data Test', len(x_test))
```

Modelling.
Disini akan menggunakan beberapa model algoritma

LARS

```python
from sklearn import linear_model
```

```python
lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train, y_train)
```

```python
pred_lars = lars.predict(x_test)
```

Evaluasi model

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
```

```python

mae_lars = mean_absolute_error(y_test, pred_lars)
mse_lars = root_mean_squared_error(y_test, pred_lars)
r2_lars = r2_score(y_test, pred_lars)

print(f'MAE : {mae_lars}')
print(f'MSE : {mse_lars}')
print(f'R2 : {r2_lars}')
```

```python
# dataframe dari evaluasi model
data = {
    'MAE' : [mae_lars],
    'MSE' : [mse_lars],
    'R2': [r2_lars]
}

df_results = pd.DataFrame(data, index=['Lars'])
df_results
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression().fit(x_train, y_train)
```

```python
pred_LR = LR.predict(x_test)
```

```python
mae_LR = mean_absolute_error(y_test, pred_LR)
mse_LR = root_mean_squared_error(y_test, pred_LR)
r2_LR = r2_score(y_test, pred_LR)

print(f'MAE : {mae_LR}')
print(f'MSE : {mse_LR}')
print(f'R2 : {r2_LR}')
```

```python
# memasukkan ke dataframe
df_results.loc['Linear Regression'] = [mae_LR, mse_LR, r2_LR]
df_results
```

GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
GBR = GradientBoostingRegressor(random_state=100)
GBR.fit(x_train, y_train)
```

```python
pred_GBR = GBR.predict(x_test)
```

```python
mae_GBR = mean_absolute_error(y_test, pred_GBR)
mse_GBR = root_mean_squared_error(y_test, pred_GBR)
r2_GBR = r2_score(y_test, pred_GBR)

print(f'MAE : {mae_GBR}')
print(f'MSE : {mse_GBR}')
print(f'R2 : {r2_GBR}')
```

```python
# memasukkan ke dataframe
df_results.loc['GradientBoostingRegressor'] = [mae_GBR, mse_GBR, r2_GBR]
df_results
```
